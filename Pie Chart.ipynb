```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
plt.pie(c1)
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
plt.pie(c1,labels =classes)
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
plt.pie(c1,labels =classes,explode =[0,0.1,0,0,0])
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
plt.pie(c1,labels =classes,explode =[0,0.1,0.001,0,0.05])
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
c =['r','g','b','y','m']
plt.pie(c1,labels =classes,explode =[0,0.1,0,0,0],colors = c)
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
c =['r','g','b','y','m']
plt.pie(c1,labels =classes,explode =[0,0.1,0,0,0],colors = c,autopct = '%0.1f%%',shadow =True)
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
c =['r','g','b','y','m']
plt.pie(c1,labels =classes,explode =[0,0.1,0,0,0],colors = c,autopct = '%0.1f%%',shadow =True,radius =1.414,startangle =90)
plt.show()
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,15,20,25,10]
c =['r','g','b','y','m']
text ={'fontsize':15}
plt.pie(c1,labels =classes,explode =[0,0.1,0,0,0],colors = c,autopct = '%0.1f%%',shadow =True,radius =1.414,startangle =90,textprops =text)
plt.show()
```
