```python
df = pd.read_csv('C:\\Users\\Nikhil Gupta\\Downloads\\archive\\googleplaystore.csv',nrows =1000)
df
```

```python
x = df["Rating"]
y = df["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize =(16,9))
plt.scatter(x,y,c ="teal")
plt.title("Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize =(16,9))
plt.scatter(x,y,c ="teal",marker = '*',s =100)
plt.title("Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize =(16,9))
plt.scatter(x,y,c ="teal",marker = '*',s =100,alpha = 0.5)
plt.title("Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize =(16,9))
plt.scatter(x,y,c ="teal",marker = '*',s =150,alpha = 0.5,linewidth = 2,edgecolor = 'black')
plt.title("Scatter Plot")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```
