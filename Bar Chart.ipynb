```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes =['Python','R','AI','ML','DS']
c1 =[30,10,20,25,10]
c2 =[40,5,20,20,10]
c3 =[35,5,30,15,15]
```

```python
plt.bar(classes,c1)
```

```python
plt.barh(classes,c1)
```

```python
plt.bar(classes,c1,width =0.8)
```

```python
plt.bar(classes,c1,width =0.8,align ='edge',color ='m')
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='m',edgecolor ='black')
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='m',edgecolor ='black',linewidth =2)
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9)
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,linestyle ='--')
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.legend()
```

```python
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,visible =False)
```

```python
style.use("ggplot")
plt.figure(figsize =(16,9))
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize =(16,9))
plt.bar(classes,c1,width =0.7,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.legend()
plt.title("Bar Chart",fontsize =25)
plt.xlabel("Classes",fontsize =20)
plt.ylabel("Students",fontsize =20)
```

```python
style.use("ggplot")
plt.figure(figsize =(16,9))
plt.bar(classes,c1,width =0.4,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.bar(classes,c2,width =0.4,align ='center',color ='purple',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class2 Students')
plt.legend()
plt.title("Bar Chart",fontsize =25)
plt.xlabel("Classes",fontsize =20)
plt.ylabel("Students",fontsize =20)
```

```python
style.use("ggplot")
classes_index =np.arange(len(classes))
width =0.4
plt.figure(figsize =(16,9))
plt.bar(classes_index,c1,width,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.bar(classes_index+width,c2,width,align ='center',color ='purple',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class2 Students')
plt.legend()
plt.title("Bar Chart",fontsize =25)
plt.xlabel("Classes",fontsize =20)
plt.ylabel("Students",fontsize =20)
```

```python
style.use("ggplot")
classes_index =np.arange(len(classes))
width =0.4
plt.figure(figsize =(16,9))
plt.bar(classes_index,c1,width,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.bar(classes_index+width,c2,width,align ='center',color ='purple',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class2 Students')
plt.legend()
plt.title("Bar Chart",fontsize =25)
plt.xlabel("Classes",fontsize =20)
plt.ylabel("Students",fontsize =20)
plt.xticks(classes_index+width/2,classes)
```

```python
style.use("ggplot")
classes_index =np.arange(len(classes))
width =0.2
plt.figure(figsize =(16,9))
plt.bar(classes_index,c1,width,align ='center',color ='teal',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class1 Students')
plt.bar(classes_index+width,c2,width,align ='center',color ='crimson',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class2 Students')
plt.bar(classes_index+2*width,c2,width,align ='center',color ='olivedrab',edgecolor ='black',linewidth =2,alpha =0.9,label ='Class3 Students')
plt.legend()
plt.title("Bar Chart",fontsize =25)
plt.xlabel("Classes",fontsize =20)
plt.ylabel("Students",fontsize =20)
plt.xticks(classes_index+width,classes)
```
