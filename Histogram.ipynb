```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45,(100))
phy_students_age =np.random.randint(18,50,(100))
print(ml_students_age)
print(phy_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("ml_students_age")
plt.show()
```

```python
plt.hist(ml_students_age)
plt.title("ml_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins)
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins = 10,rwidth =0.8)
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins,rwidth =0.8,histtype ='barstacked')
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins,rwidth =0.8,histtype ='barstacked',align ='right',orientation ='horizontal')
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins,rwidth =0.8,histtype ='barstacked',color ='m')
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist(phy_students_age,bins,rwidth =0.8,histtype ='barstacked')
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.legend("ML Students",loc =2)
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.hist([ml_students_age,phy_students_age],bins,rwidth =0.8,histtype ='bar'
        ,color =['m','y'])
plt.title("Phy_students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.figure(figsize =(16,9))
plt.hist([ml_students_age,phy_students_age],bins,rwidth =0.8,histtype ='bar'
        ,color =['m','y'],label =["ML Students","Phy Students"])
plt.title("students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.legend()
plt.show()
```

```python
bins =[15,20,25,30,35,40,45,50]
plt.figure(figsize =(14,9))
plt.hist([ml_students_age,phy_students_age],bins,rwidth =0.8,histtype ='bar'
        ,color =['m','y'],label =["ML Students","Phy Students"])
plt.title("students_age")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.legend()
plt.show()
```
